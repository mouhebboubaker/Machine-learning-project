```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer
import re 

nltk.download('stopwords')  # Ensure stop words are available
# Define Arabic stop words
stop_words = set(stopwords.words("arabic"))
stemmer=ISRIStemmer()

def process_quotes(text):
    tokens=text.split()
    tokens=[ x for x in tokens if x not in stop_words]
    tokens=[stemmer.stem(word) for word in tokens if word]
    return  ' '.join(tokens)

prepared_df=df.copy()
prepared_df["quote"]=df["quote"].apply(process_quotes)
from IPython.display import display
display(prepared_df)
# prepared_df.to_csv('filtred.csv',index=True,encoding='utf-8-sig')

```

Vectorization using tf-IDF


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#transforme targets into numeric variable
label_encoder=LabelEncoder()
prepared_df["classe_num"]=label_encoder.fit_transform(prepared_df["classe"])
 
#vecotrozation of quotes
v = TfidfVectorizer()
x = v.fit_transform(prepared_df["quote"])
y=prepared_df["classe_num"]
# print(v.vocabulary_)
all_feature_names=v.get_feature_names_out()
# print(len(all_feature_names)) # we have 8082 different word in our quotes

#affichage de 20 idf of words to see if the vectorization is right
# for word in all_feature_names :
#     index=v.vocabulary_[word]
#     idf_score=v.idf_[index]
#     print(f"{word} : { idf_score}")

```

```python
#using knn 

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

#Random Baseline Accuracy:
#if we don't use a model the probability of correct classification is :
# if we randomly assign one of the 99 classes to a quote, the probability of assiging the correct class is : 
# 1/ 99 = 0.01 %



k_values = range(1, 500,25)  # Test values of k from 1 to 20
best_k = None
best_accuracy = 0

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x, y, cv=5, scoring='accuracy')  # 5-fold CV
    avg_score = scores.mean()
    print(f"k={k}, Cross-Validation Accuracy: {avg_score:.4f}")
    if avg_score > best_accuracy:
        best_k = k
        best_accuracy = avg_score
print(f"\nBest k: {best_k} with Cross-Validation Accuracy: {best_accuracy:.4f}")


 
```
